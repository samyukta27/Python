# List:
- one of the data structures in python
- Data structure means to store,sort and search Options
- its one common data structure in python,the items of list seperated by comma and enclosed by square brackets

###example
* list1=[1,2,3,4]
* list2==["gitam",1,2,3]

```
li = [1,8,16,9,2]
print(li)
print(li[2])
print(li[-3])
print(li[2:])
print(li[2:4])
print(li[-3:-1])
```

```
li = ["Gitam",1234,1235,1236,"Hyd"]
print(li)
print(li[0][2])
```

```
# delete
del li[3]
print(li)
```

```
# update
li[1] = 123
print(li)
```

```
# basic list functions
li = [1,9,116,25,5]
print(max(li))
print(min(li))
print(sum(li))
print(len(li))
print(sum(li)//len(li))
```

```
li = [1,9,116,25,5,9]
li.append(89)
li.insert(2,56)
li.count(9)
li.index(5)
li.sort()
li.pop()
li.pop(1)
li
```

### del-delete the list item from the list
#### pop-remove last index value
##### pop(index)-removes the items from specified index

```
li = [1,9,116,25,5,9]
lil = [76,78,79]
li.extend(lil)
li.reverse()
li
```

```
# fuction to return the large number from given list
def smallnumber(li):
  li.sort()
  return li[0]
def genericsmall(li,n):
  li.sort()
  return li[n]
li = [19,1,25,45,76]
print(smallnumber(li))
print(genericsmall(li,3))
```

# String Built in Methods:
- upper

```
s1 = 'Python'
print(s1.upper())
print(s1.lower())
print(s1.isupper())
print(s1.islower())
print(s1.istitle())
print(s1.isnumeric())
print(s1.isalpha())
```
